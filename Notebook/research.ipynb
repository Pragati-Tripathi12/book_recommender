```python
import pandas as pd
import numpy as np
```

```python
books=pd.read_csv("BX-Books.csv",sep=";",encoding="latin-1",error_bad_lines=False)
```

```python
books.head()
```

```python
books.iloc[10]['Image-URL-M']
```

```python
books.shape
```

```python
books.columns
```

```python
books=books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-L']]
books.head()
```

```python
books.rename(columns={'Book-Title':'title','Book-Author':'author','Year-Of-Publication':'year','Publisher':'publisher','Image-URL-L':'image'},inplace=True)
books.head()
```

```python
users=pd.read_csv("BX-Users.csv",sep=";",encoding="latin-1",error_bad_lines=False)
users.head()
```

```python
users.shape
```

```python
users.rename(columns={'User-ID':'user_id','Location':'location','Age':'age'},inplace=True)
users.head()
```

```python
ratings=pd.read_csv("BX-Book-Ratings.csv",sep=";",encoding="latin-1",error_bad_lines=False)
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
ratings.head()
```

```python
print(books.shape,users.shape,ratings.shape)
```

```python
ratings['user_id'].value_counts()
```

```python
x=ratings['user_id'].value_counts()>200 
x
```

```python
x[x].shape
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
ratings.shape
```

```python
ratings.head
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
number_ratings=ratings_with_books.groupby('title')['rating'].count().reset_index()
number_ratings.head()
```

```python
number_ratings.rename(columns={'rating':'num_of_ratings'},inplace=True )
number_ratings.head()
```

```python
final_ratings=ratings_with_books.merge(number_ratings,on='title')
final_ratings.head()
```

```python
final_ratings.shape
```

```python
final_ratings=final_ratings[final_ratings['num_of_ratings'] >=50]
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id','title'],inplace=True)
final_ratings.shape
```

```python
final_ratings.head()
```

```python
final_ratings.isnull().sum()
```

```python
final_ratings.dropna(inplace=True)
final_ratings.isnull().sum()
```

```python
book_pivot=final_ratings.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse_matrix=csr_matrix(book_pivot)
book_sparse_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
book_pivot.iloc[237,:]
```

```python
model.fit(book_sparse_matrix)
```

```python
distance,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=5)
```

```python
distance
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```
